```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Dataset.csv')
```

```python
proporcao = df.groupby(['Genero', 'Estado Civil']).size().reset_index(name = 'Proporção')

proporcao['Proporção'] = proporcao['Proporção']/proporcao['Proporção'].sum()*100

print(proporcao)
```

```python
pivot = proporcao.pivot(index='Genero',
                        columns='Estado Civil',
                        values='Proporção')

pivot.plot(kind='bar',
           stacked='true')

plt.title('Proporção de Clientes por Gênero e Estado Civil')
plt.xlabel('Gênero')
plt.ylabel('Proporção')
plt.legend(title='Estado Civil')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
# Mostrando o gráfico
plt.show()
```
