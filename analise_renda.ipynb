```python
import pandas as pd 
```

```python
df = pd.read_csv('Dataset.csv')
```

```python
#criando dicionário para renda média
renda_map = {
    'R$ 2.001 - R$ 4.000': 3000,
    'R$ 4.001 - R$ 6.000': 5000,
    'R$ 6.001 - R$ 8.000': 7000,
    'R$ 8.001 - R$ 10.000': 9000,
    'Acima de R$ 10.000': 11000
}

#Add novo campo de renda média, mas com valores inteiros 
df['renda_media_numerica'] = df['Renda Média'].map(renda_map)
```

```python
media_renda = df['renda_media_numerica'].mean()
print(media_renda)

menor_renda_media = df[df['renda_media_numerica'] < media_renda]


# Filtrando valores maiores ou iguais a 3000
maior_renda_media = df[df['renda_media_numerica'] > media_renda]

```
